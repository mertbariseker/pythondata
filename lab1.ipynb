```python
def is_prime(num):
    """Helper function to check if a number is prime."""
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def task_9(*args):
    """Filters the entered list to leave only non-prime numbers in descending order."""
    non_primes = [n for n in args if not is_prime(n)]
    return sorted(non_primes, reverse=True)
```

```python
print(task_9(3, 12, 15, 7, 11, 35, 18))  # Expected output: [35, 18, 15, 12]
```

```python
print(task_9(0, 34, 2, 1, 15, 22, 4, 23, 6))  # Expected output: [34, 22, 15, 6, 4]
```
